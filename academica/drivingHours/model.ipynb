```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
# Using the given image to prepare the dataset
data = {
    'Hours' : [10, 9, 2, 10, 15, 16, 11, 16],
    'Risk_Score' : [95, 80, 10, 45, 50, 98, 38, 93]
}
```

```python
# Preparing dataframe of the given data
df = pd.DataFrame(data)
```

```python
# Prepare data for model
# Divide the data into input and output variables
X = df[['Hours']]
y = df[['Risk_Score']]
```

```python
# create a linear regression model
model = LinearRegression()
model.fit(X, y)
```

```python
# get coefficients
slope = model.coef_[0]
intercept = model.intercept_
```

```python
# equation of the line
print(f"y = {slope.item():.2f}x + {intercept.item():.2f}")
```

```python
# predict the values
y_pred = model.predict(X)
```

```python
# Calulate the mean squared error
mse = mean_squared_error(y, y_pred)
print(f"{mse:.2f}")
```

```python
# plot the best fitline and the actual data
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, y_pred, color='red', label='Best fit line')
plt.xlabel('Hrs spent driving')
plt.ylabel('Risk Score')
plt.title('Simple Linear Regression')
plt.legend()
plt.grid(True)
plt.show()
```
